### Bilinear Interpolation
Let's write our own function for resizing the image using bilinear interpolation

```python
def resize_image(img:np.ndarray, fx=1, fy=1):
	height, width = img.shape
	resized_height, resized_width = round(height*fx), round(width*fy)
	x_ratio = (height)/(resized_height) if resized_height != 0 else 0
	y_ratio = (width)/(resized_width) if resized_width != 0 else 0
	resized_image = np.empty((resized_height, resized_width), dtype=int)
	for i in range(resized_height):
		for j in range(resized_width):
			x_weight = x_ratio * (i)
			y_weight = y_ratio * (j)
			x_low, x_high = math.floor(x_weight), min(math.ceil(x_weight), height-1)
			y_low, y_high = math.floor(y_weight), min(math.ceil(y_weight), width -1)
			if (x_low == x_high) and (y_low == y_high):
				pixel = img[x_low, y_low]
			elif x_low == x_high:
				p1, p2 = img[x_low, y_low], img[x_low, y_high]
				pixel = p1*(y_high - y_weight) + p2*(y_weight - y_low)
			elif y_low == y_high:
				p1, p2 = img[x_low, y_low], img[x_high, y_low]
				pixel = p1*(x_high - x_weight) + p2*(x_weight - x_low)
			else:
				a = img[x_low][y_low]
				b = img[x_high][y_low]
				c = img[x_low][y_high]
				d = img[x_high][y_high]
				p1 = a*(x_high - x_weight) + b*(x_weight - x_low)
				p2 = c*(x_high - x_weight) + d*(x_weight - x_low)
				pixel = p1*(y_high - y_weight) + p2*(y_weight - y_low)
			resized_image[i, j] = int(pixel)
	return resized_image
```

### Plotting and comparison with inbuilt function

Let's resize the images using both inbuilt and our own defined function for bilinear interpolation and plot them side by side to compare with different scale factors, 0.5, 1 and 2.

```python
factor = 1
resized_image_user = resize_image(gray_img, fx=factor, fy=factor)
resized_image_inbuilt = cv2.resize(gray_img, None, fx=factor, fy=factor, interpolation=cv2.INTER_LINEAR_EXACT)
size = len(resized_image_inbuilt)

fig, axs = plt.subplots(1, 2)
fig.suptitle(f"Resizing using scale factor = {factor} ({size}x{size})")
axs[0].imshow(resized_image_user, cmap="gray")
axs[0].set_title("User defined")
axs[1].imshow(resized_image_inbuilt, cmap="gray")
axs[1].set_title("Inbuilt")
plt.show()

compare = (resized_image_user == resized_image_inbuilt).sum()
print(f"Pixel wise comparison, similar number of pixels = {compare} in total of {size*size}")
```

```python
factor = 2
resized_image_user = resize_image(gray_img, fx=factor, fy=factor)
resized_image_inbuilt = cv2.resize(gray_img, None, fx=factor, fy=factor, interpolation=cv2.INTER_LINEAR_EXACT)
size = len(resized_image_inbuilt)

fig, axs = plt.subplots(1, 2)
fig.suptitle(f"Resizing using scale factor = {factor} ({size}x{size})")
axs[0].imshow(resized_image_user, cmap="gray")
axs[0].set_title("User defined")
axs[1].imshow(resized_image_inbuilt, cmap="gray")
axs[1].set_title("Inbuilt")
plt.show()

compare = (resized_image_user == resized_image_inbuilt).sum()
print(f"Pixel wise comparison, similar number of pixels = {compare} in total of {size*size}")
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
input_image = cv2.imread('Nagasai.jpg')
plt.imshow(cv2.cvtColor(input_image,cv2.COLOR_BGR2RGB))
```

```python
img = cv2.cvtColor(input_image,cv2.COLOR_BGR2GRAY)
plt.imshow(img,cmap='gray')
```

```python
canny_edged = cv2.Canny(img, 50,200)
plt.imshow(canny_edged,cmap="gray")
```

```python
sigma = 1
log_edged = cv2.GaussianBlur(img, (0, 0), sigmaX=sigma)
log_edged = cv2.Laplacian(log_edged, cv2.CV_64F)
log_edged = np.uint8(np.absolute(log_edged))
plt.imshow(log_edged,cmap="gray")
```

```python
plt.figure(figsize=(10,10))
c = 100
sharpened_canny = cv2.addWeighted(img, 1, np.zeros_like(canny_edged), 0, c)
sharpened_log = cv2.addWeighted(img, 1, np.zeros_like(log_edged), 0, c)
plt.subplot(121)
plt.imshow(sharpened_canny,cmap = "gray")
plt.subplot(122)
plt.imshow(sharpened_log,cmap = "gray")
```
